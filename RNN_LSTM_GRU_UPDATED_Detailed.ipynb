```python
# ============================
# NIFTY Price Prediction Teaching Notebook
# RNN / LSTM / GRU — many variants, visuals & comparisons
# Simplified version without functions for easier teaching
# ============================

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time, math
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.seasonal import seasonal_decompose
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, SimpleRNN, LSTM, GRU
from tensorflow.keras.callbacks import EarlyStopping

# -----------------------
# Reproducibility
# -----------------------
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# -----------------------
# Configuration
# -----------------------
TICKER = "^NSEI"
START_DATE = "2010-01-01"
END_DATE = None
USE_FEATURES = ["Close"]
TIME_STEPS = 60
TEST_SPLIT = 0.2
VIZ_SAMPLE_DAYS = 1000
EPOCHS = 20
BATCH_SIZE = 32

# -----------------------
# 1) Download Data
# -----------------------
print("Downloading data for", TICKER)
df = yf.download(TICKER, start=START_DATE, end=END_DATE, progress=False)
if df.empty:
    raise RuntimeError("No data downloaded. Check ticker or date range.")
df = df.reset_index()
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)
print("Data downloaded:", df.shape)
print(df.head())

# -----------------------
# Visualize Data
# -----------------------
plt.figure(figsize=(14,5))
plt.plot(df['Date'], df['Close'], lw=1)
plt.title("NIFTY — Closing Price (full range)")
plt.xlabel("Date"); plt.ylabel("Close Price")
plt.show()

plt.figure(figsize=(14,5))
plt.plot(df['Date'].iloc[-VIZ_SAMPLE_DAYS:], df['Close'].iloc[-VIZ_SAMPLE_DAYS:], lw=1)
plt.title(f"NIFTY — Closing Price (last {VIZ_SAMPLE_DAYS} days)")
plt.xlabel("Date"); plt.ylabel("Close Price")
plt.show()

# -----------------------
# 2) Seasonal Decomposition
# -----------------------
try:
    decomp_period = 252
    decomposed = seasonal_decompose(df['Close'], period=decomp_period, model='multiplicative', extrapolate_trend='freq')
    fig = decomposed.plot()
    fig.set_size_inches(14,10)
    plt.suptitle("Seasonal Decompose (multiplicative)")
    plt.show()
except Exception as e:
    print("Seasonal decomposition skipped:", e)

# -----------------------
# 3) Preprocessing
# -----------------------
features = USE_FEATURES.copy()
data = df[features].astype('float32').values
print("Using features:", features)

scaler = MinMaxScaler(feature_range=(0,1))
data_scaled = scaler.fit_transform(data)
print("Scaled data shape:", data_scaled.shape)

# -----------------------
# 4) Create Sequences
# -----------------------
X, y = [], []
for i in range(TIME_STEPS, len(data_scaled)):
    X.append(data_scaled[i-TIME_STEPS:i])
    y.append(data_scaled[i])
X = np.array(X)
y = np.array(y)
print("Sequence shapes X, y:", X.shape, y.shape)

# Show example window
sample_idx = -100
example_window = scaler.inverse_transform(X[sample_idx].reshape(-1, X.shape[-1])).flatten()[:10]
print("\nExample window (first 10 values in original scale):")
print(example_window)

# -----------------------
# 5) Train-test split
# -----------------------
split_idx = int((1 - TEST_SPLIT) * len(X))
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# -----------------------
# 6) Model Building - Explicit and Repetitive
# -----------------------
n_features = X.shape[-1]
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# RNN Models
print("\n" + "="*50)
print("Building RNN-32")
model_rnn_32 = Sequential()
model_rnn_32.add(SimpleRNN(32, input_shape=(TIME_STEPS, n_features)))
model_rnn_32.add(Dropout(0.1))
model_rnn_32.add(Dense(n_features))
model_rnn_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_rnn_32.summary())

print("\n" + "="*50)
print("Building RNN-64-32")
model_rnn_64_32 = Sequential()
model_rnn_64_32.add(SimpleRNN(64, return_sequences=True, input_shape=(TIME_STEPS, n_features)))
model_rnn_64_32.add(Dropout(0.2))
model_rnn_64_32.add(SimpleRNN(32))
model_rnn_64_32.add(Dropout(0.2))
model_rnn_64_32.add(Dense(n_features))
model_rnn_64_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_rnn_64_32.summary())

print("\n" + "="*50)
print("Building RNN-128-64-32")
model_rnn_128_64_32 = Sequential()
model_rnn_128_64_32.add(SimpleRNN(128, return_sequences=True, input_shape=(TIME_STEPS, n_features)))
model_rnn_128_64_32.add(Dropout(0.3))
model_rnn_128_64_32.add(SimpleRNN(64, return_sequences=True))
model_rnn_128_64_32.add(Dropout(0.3))
model_rnn_128_64_32.add(SimpleRNN(32))
model_rnn_128_64_32.add(Dropout(0.3))
model_rnn_128_64_32.add(Dense(n_features))
model_rnn_128_64_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_rnn_128_64_32.summary())

# LSTM Models
print("\n" + "="*50)
print("Building LSTM-32")
model_lstm_32 = Sequential()
model_lstm_32.add(LSTM(32, input_shape=(TIME_STEPS, n_features)))
model_lstm_32.add(Dropout(0.1))
model_lstm_32.add(Dense(n_features))
model_lstm_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_lstm_32.summary())

print("\n" + "="*50)
print("Building LSTM-64-32")
model_lstm_64_32 = Sequential()
model_lstm_64_32.add(LSTM(64, return_sequences=True, input_shape=(TIME_STEPS, n_features)))
model_lstm_64_32.add(Dropout(0.2))
model_lstm_64_32.add(LSTM(32))
model_lstm_64_32.add(Dropout(0.2))
model_lstm_64_32.add(Dense(n_features))
model_lstm_64_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_lstm_64_32.summary())

print("\n" + "="*50)
print("Building LSTM-128-64-32")
model_lstm_128_64_32 = Sequential()
model_lstm_128_64_32.add(LSTM(128, return_sequences=True, input_shape=(TIME_STEPS, n_features)))
model_lstm_128_64_32.add(Dropout(0.3))
model_lstm_128_64_32.add(LSTM(64, return_sequences=True))
model_lstm_128_64_32.add(Dropout(0.3))
model_lstm_128_64_32.add(LSTM(32))
model_lstm_128_64_32.add(Dropout(0.3))
model_lstm_128_64_32.add(Dense(n_features))
model_lstm_128_64_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_lstm_128_64_32.summary())

# GRU Models
print("\n" + "="*50)
print("Building GRU-32")
model_gru_32 = Sequential()
model_gru_32.add(GRU(32, input_shape=(TIME_STEPS, n_features)))
model_gru_32.add(Dropout(0.1))
model_gru_32.add(Dense(n_features))
model_gru_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_gru_32.summary())

print("\n" + "="*50)
print("Building GRU-64-32")
model_gru_64_32 = Sequential()
model_gru_64_32.add(GRU(64, return_sequences=True, input_shape=(TIME_STEPS, n_features)))
model_gru_64_32.add(Dropout(0.2))
model_gru_64_32.add(GRU(32))
model_gru_64_32.add(Dropout(0.2))
model_gru_64_32.add(Dense(n_features))
model_gru_64_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_gru_64_32.summary())

print("\n" + "="*50)
print("Building GRU-128-64-32")
model_gru_128_64_32 = Sequential()
model_gru_128_64_32.add(GRU(128, return_sequences=True, input_shape=(TIME_STEPS, n_features)))
model_gru_128_64_32.add(Dropout(0.3))
model_gru_128_64_32.add(GRU(64, return_sequences=True))
model_gru_128_64_32.add(Dropout(0.3))
model_gru_128_64_32.add(GRU(32))
model_gru_128_64_32.add(Dropout(0.3))
model_gru_128_64_32.add(Dense(n_features))
model_gru_128_64_32.compile(optimizer='adam', loss='mse', metrics=['mae'])
print(model_gru_128_64_32.summary())

# -----------------------
# 7) Train all models
# -----------------------
models = {
    "RNN-32": model_rnn_32,
    "RNN-64-32": model_rnn_64_32,
    "RNN-128-64-32": model_rnn_128_64_32,
    "LSTM-32": model_lstm_32,
    "LSTM-64-32": model_lstm_64_32,
    "LSTM-128-64-32": model_lstm_128_64_32,
    "GRU-32": model_gru_32,
    "GRU-64-32": model_gru_64_32,
    "GRU-128-64-32": model_gru_128_64_32
}

history = {}
results = {}
timings = {}
predictions = {}

for name, model in models.items():
    print("\n" + "="*50)
    print("Training model:", name)

    start = time.time()
    h = model.fit(
        X_train, y_train,
        validation_split=0.15,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[early_stop],
        verbose=1
    )
    elapsed = time.time() - start

    # Predict and evaluate
    y_pred_scaled = model.predict(X_test)
    y_test_rescaled = scaler.inverse_transform(y_test)
    y_pred_rescaled = scaler.inverse_transform(y_pred_scaled)

    close_idx = features.index("Close") if "Close" in features else 0
    rmse = math.sqrt(mean_squared_error(y_test_rescaled[:, close_idx], y_pred_rescaled[:, close_idx]))
    mae = mean_absolute_error(y_test_rescaled[:, close_idx], y_pred_rescaled[:, close_idx])

    print(f"{name} RMSE (Close): {rmse:.4f}, MAE (Close): {mae:.4f}, Time: {elapsed:.1f}s")

    history[name] = h
    results[name] = {"rmse": rmse, "mae": mae}
    timings[name] = elapsed
    predictions[name] = {"y_true": y_test_rescaled[:, close_idx], "y_pred": y_pred_rescaled[:, close_idx]}

# -----------------------
# 8) Visual Comparisons
# -----------------------
plt.figure(figsize=(14,6))
for name, h in history.items():
    if "val_loss" in h.history:
        plt.plot(h.history['val_loss'], label=name)
plt.title("Validation Loss (MSE) per Model Variant")
plt.xlabel("Epoch")
plt.ylabel("Val Loss")
plt.legend()
plt.show()

plt.figure(figsize=(14,6))
for name, h in history.items():
    if "val_mae" in h.history:
        plt.plot(h.history['val_mae'], label=name)
plt.title("Validation MAE per Model Variant")
plt.xlabel("Epoch")
plt.ylabel("Val MAE")
plt.legend()
plt.show()

# Runtime comparison
plt.figure(figsize=(10,6))
names = list(timings.keys())
times = [timings[n] for n in names]
sns.barplot(x=times, y=names)
plt.xlabel("Training Time (s)")
plt.title("Training Time per Model Variant")
plt.show()

# -----------------------
# 9) Show predictions for best model
# -----------------------
best_name = min(results.keys(), key=lambda k: results[k]['rmse'])
print("\nBest model by RMSE (Close):", best_name, results[best_name])

y_true = predictions[best_name]['y_true']
y_pred = predictions[best_name]['y_pred']

plt.figure(figsize=(14,6))
plt.plot(y_true, label='Actual Close (Test)')
plt.plot(y_pred, label='Predicted Close (Test)')
plt.title(f"Actual vs Predicted Close Prices — {best_name}")
plt.xlabel("Test sample index")
plt.ylabel("Price")
plt.legend()
plt.show()

# Zoom view
zoom_n = min(200, len(y_true))
plt.figure(figsize=(14,6))
plt.plot(y_true[-zoom_n:], lw=1.5, label='Actual (last points)')
plt.plot(y_pred[-zoom_n:], lw=1.2, label='Predicted (last points)')
plt.title(f"Zoom (last {zoom_n}) — {best_name}")
plt.legend()
plt.show()

# -----------------------
# 10) Error analysis
# -----------------------
residuals = y_true - y_pred
plt.figure(figsize=(10,4))
sns.histplot(residuals, bins=50, kde=True)
plt.title(f"Residual Distribution — {best_name}")
plt.xlabel("Residual (Actual - Predicted)")
plt.show()

# -----------------------
# 11) Summary table
# -----------------------
summary_df = pd.DataFrame([
    {"Model": name, "RMSE": results[name]['rmse'], "MAE": results[name]['mae'], "Training Time (s)": round(timings[name],1)}
    for name in results
]).sort_values("RMSE")
print("\n=== Summary Table (sorted by RMSE) ===")
print(summary_df)

# -----------------------
# Teaching Notes
# -----------------------
print("""
Teaching progression notes:
1) Start with raw plot + decomposition -> motivate stationarity/trend/seasonality.
2) Show the effect of scaling and why sequences are needed.
3) Train small RNN/LSTM/GRU (32 units) — quick results to show model shapes.
4) Increase depth & units (64->32, 128->64->32) and discuss capacity vs overfitting.
5) Show validation loss curves and explain EarlyStopping.
6) Show runtime differences (RNN < GRU < LSTM typically).
7) Demonstrate forecasting quality (RMSE/MAE) and error/residual analysis.

Key concepts to explain:
- Why we use sequential data (time steps)
- How RNN/LSTM/GRU differ in architecture
- The role of dropout in preventing overfitting
- How scaling helps with gradient stability
- Why we use early stopping
- How to interpret RMSE and MAE in context of stock prices
""")
```
